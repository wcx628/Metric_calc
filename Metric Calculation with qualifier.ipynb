```python
def calc_child_death_ratio(df):
    '''Main metric calculator'''
    return (df['Sons who have died'].sum() + df['Daughters who have died'].sum()) / df['Total children ever born'].sum()
```

```python
def cb_gen(n):
    '''potential combination generator, 1111111 indicates all categories included, 
    1100111 indicates category 2 and 3 not included, potential category combination
    stored in queue to determine whether to go deeper'''
    
    q = ['1'*n]
    visited = set(q)  # to eliminate potential double visit from top down traverse
    i,j = -1,0
    while i<j:
        i += 1
        cur = q[i]
        # print(q)
        t_flag = (yield cur)  # wait for outside signal for whether to traverse the subtree
        if t_flag:
            for k in range(n):
                if cur[k] == '1':
                    newstr = cur[:k]+'0'+cur[k+1:]
                    if newstr not in visited:
                        q.append(newstr)
                        visited.add(newstr)
                        j += 1
        else:
            pass
```

```python
def filter_calc(df, dim, filter_list):
    '''Helper function to generate filtered view of df'''
    ft = pd.Series([True]*df.shape[0], df.index)
    for idx,dm in enumerate(dim):
        dm_set = df[dm].unique()
        dm_set = [dm_set[i] for i in range(len(dm_set)) if filter_list[idx][i] == '1']
        ft = ft & (df[dm].isin(dm_set))
        
    return df.loc[ft]


def qualifier(df, row_threshold):
    '''Main qualifier to determine whether the subset qualifies'''
    
    return df.shape[0] >= row_threshold


        
def calc_all_analytics(df, dim, row_threshold, metric_calc):
    '''Main looper to calculate all qualified metrics
    
    Input:
      df: dataframe to calculate all the metrics from
      dim: name of the dimension columns in the dataframe
      row_threshold: the threshold for which the number of row should exceed
      metric_calc: the metric calculator to calculate metrics for all the combination
      
    '''
    
    res = {}
    
    dim_l = 0  # level of the dimension
    filter_list = []  # store all the dimension filter as a stack
    gen_list = []   # store all the generators for each dimension
    dim_count = [ len(df[dim[i]].unique()) for i in range(len(dim)) ]  # count of each dimension
    
    t_flag = True  # track whether the current combination qualifies
    just_start = True  # handle initialization of the generator
    
    while dim_l < len(dim) and dim_l >= 0:
        
        if dim_l == len(gen_list):    # traverse deeper into next level of dimension
            gen_list.append( cb_gen(dim_count[dim_l]) )
            dim_s = next(gen_list[dim_l])
            just_start = True
            t_flag = True
        else:
            just_start = False
            
        try:                      # main looping, send signal back to generator, back trace
            if not just_start:
                dim_s = gen_list[dim_l].send(t_flag)
            filter_list.append(dim_s)
            # print(filter_list, dim_s, gen_list)
            
            df_sub = filter_calc(df, dim[:dim_l+1], filter_list[:dim_l+1]) # get the subset from all the dimension filter
            
            if qualifier(df_sub, row_threshold):    # check the qualifier to determine whether to go to the next dimension
                res[tuple(filter_list)] = metric_calc(df_sub)  # store the metric
                if dim_l < len(dim) - 1:
                    dim_l += 1    # go to next dimension if there is more
                else:
                    _ = filter_list.pop()  # remove the current filter for the next filter if stays at current dimension

            else:
                _ = filter_list.pop()
                t_flag = False   # if current subset doesn't qualify, back trace and send the signal to generator
                
        except StopIteration:
            _ = gen_list.pop()
            if filter_list:      # handle back trace of filter list, since it could be popped from step above
                _ = filter_list.pop()
            dim_l -= 1           # finish current dimension loop, back trace to previous dimension
            t_flag = True
            # print(dim_l, filter_list, gen_list)
            
    return res  # Right now the result is stored with tuples to indicate the dimension filter, 
                # it can be modified to include more information
            
            
```

```python
calc_all_analytics(women_df, ['Month of interview', "Respondent's month of birth"], 6000, calc_child_death_ratio)
```
